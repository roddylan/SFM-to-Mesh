```python
from ultralytics import YOLO
from sfm import SFM
import os, cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = YOLO("models/yolov8n-seg.pt")
```

```python
lego_im_src = "datasets/Viking"

files = os.listdir(lego_im_src)
files.sort()
itoimg = {}
n = 0

for f in files:
    try:
        im = cv2.imread(f"{lego_im_src}/{f}")
        gim = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        itoimg[n] = f"{lego_im_src}/{f}"
        n+=1

    except:
        pass

i = 0
while True:
    frame = cv2.imread(itoimg[i])
    h, w, _ = frame.shape
    res = model(frame)
    _frame = res[0].plot()
    i+=1

    cv2.imshow("mask", _frame)
    cv2.waitKey(50)

    # fix for notebook
    if i == n:
        cv2.destroyAllWindows()
        break

# mask = res[0].masks.data
# mask = (mask.numpy()+255).reshape(w,h)
# (mask.numpy()+255).squeeze().shape
# mask = cv2.resize(mask.numpy() + 255, (w,h))
```
